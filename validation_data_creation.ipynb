## Validation Data Creation

##### * Before beginning running this bit of code, please ensure you are in the "train" directory. 
##### * Also note that this is a continuation of the previous code block so you should continue following along on the same Jupyter Notebook.

```python
# The next step would be to to seperate validation data and plot it...
# We create a folder that contains validation data...
%mkdir ../valid

# We create the same folders from the training data within the validation data
for d in directory_names:
    os.mkdir('../valid/'+d)
```

##### Below, we prompt glob to select every folder and every file ending with a ".jpg". The "*" corresponds to selecting everything.

```python
g = glob.glob('*/*.jpg')
```

```python
# We shuffle all the image files, so that we have a random order of images...
shuf = np.random.permutation(g)
# This is the first image after our random shuffle...
im = imread(shuf[0])
plt.imshow(im, cmap=cm.gray)
print(shuf[0])
plt.show()
```

#### Since we have a total of 30336 training images, it is generally a good rule of thumb to divide up the training and validation data as an 80% : 20% split, respectively.  since 20% of 30336 is 6067, that is the number of images we will transfer to the "valid" folder.

```python
for i in range(6067):
    os.rename(shuf[i], data_dir+'valid/'+shuf[i])
print("Done moving images to the validation set...")
```

#### Now, just to prove it to you, below is the code that confirms that we have indeed transfered the intented images to the validation set.

```python
numberofImages = 0
for folder in directory_names:
    for fileNameDir in os.walk(os.path.join(data_dir+"valid", folder)):
        for fileName in fileNameDir[2]:
            if fileName[-4:] != ".jpg":
              continue
            numberofImages += 1

print(numberofImages)
```
